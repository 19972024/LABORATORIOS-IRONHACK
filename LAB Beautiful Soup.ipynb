```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://www.billboard.com/charts/hot-100/"
```

```python
response=requests.get(url)
response.status_code
```

```python
#1. Realizar Web Scraping con Beautiful Soup de las 100 canciones más escuchas del momento, cogiendo tanto la canción completa como el artista. (url https://www.billboard.com/charts/hot-100/)
soup=BeautifulSoup(response.content, "html.parser")
```

```python
top_100 = soup.find_all("li", class_="o-chart-results-list__item // lrv-u-flex-grow-1 lrv-u-flex lrv-u-flex-direction-column lrv-u-justify-content-center lrv-u-border-b-1 u-border-b-0@mobile-max lrv-u-border-color-grey-light lrv-u-padding-l-050 lrv-u-padding-l-1@mobile-max")
top_100
```

```python
len(top_100)
```

```python
top100=[]
for i in top_100:
    top100.append(i.get_text())
top100
```

```python
len(top100)
```

```python
top_100_clean = []

# Iterar sobre los elementos de artista_cancion
for i in top100:
    # Dividir la cadena en partes usando el separador específico
    partes = i.split('\n\n\n\t\n\t\n\t\t\n\t\t\t\t\t')
    
    # Filtrar las partes vacías y limpiarlas
    partes_limpio = [parte.strip() for parte in partes if parte.strip()]
    
    # Unir las partes limpias con un espacio y agregarlas a la lista artista_clean
    top_100_clean.append(' '.join(partes_limpio))
    
print(top_100_clean)
```

```python
len(top_100_clean)
```

```python
titulo_cancion = []
nombre_artista = []

# Iterar sobre cada elemento de la lista
for elemento in top_100_clean:
    # Dividir el elemento en dos partes en el separador
    partes = elemento.split('\t\t\n\t\n\n\n\t\n\t')
    # Añadir la primera parte a la lista antes_separador y la segunda parte a la lista despues_separador
    titulo_cancion.append(partes[0])
    nombre_artista.append(partes[1])
    
print("Canciones:")
print(titulo_cancion)
print("\nArtistas:")
print(nombre_artista)
```

```python
#2. Guardar el scrapeo en una función. En esta función, cada vez que la ejecutes, el usuario debe:

#Introducir una canción (o parte de su título). En caso de que ese input (mínimo len=3, no case sensitive) corresponda con alguna canción, se debe devolver al usuario el nombre de la canción completa y su artista. Si se trata de esa canción, el usuario debe ser capaz de decir que lo es (con un SÍ o un YES o lo que queráis).
#En caso de encontrarla, se debe devolver "Esta canción está en el hot 100 de Billaboard".
#En caso de no encontrarla o que el usuario no confirme, se debe devolver "no hemos sido capaces de encontrar tu canción". b

def billboard_scrap():
    a = input("Introduce el título de una canción (mínimo 3 caracteres): ")
    while len(a) < 3:
        a = input("Introduce el título de una canción (mínimo 3 caracteres): ")

    url = "https://www.billboard.com/charts/hot-100/"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    top_100 = soup.find_all("li", class_="o-chart-results-list__item // lrv-u-flex-grow-1 lrv-u-flex lrv-u-flex-direction-column lrv-u-justify-content-center lrv-u-border-b-1 u-border-b-0@mobile-max lrv-u-border-color-grey-light lrv-u-padding-l-050 lrv-u-padding-l-1@mobile-max")
    top100 = [i.get_text().strip() for i in top_100]

    top_100_clean = []
    for i in top100:
        partes = i.split('\n\n\n\t\n\t\n\t\t\n\t\t\t\t\t')
        partes_limpio = [parte.strip() for parte in partes if parte.strip()]
        top_100_clean.append(' '.join(partes_limpio))
    
    titulo_cancion = []
    nombre_artista = []
    for elemento in top_100_clean:
        partes = elemento.split('\t\t\n\t\n\n\n\t\n\t')
        titulo_cancion.append(partes[0])
        nombre_artista.append(partes[1])

    cancion_top_1 = soup.find("h3", class_="c-title a-no-trucate a-font-primary-bold-s u-letter-spacing-0021 u-font-size-23@tablet lrv-u-font-size-16 u-line-height-125 u-line-height-normal@mobile-max a-truncate-ellipsis u-max-width-245 u-max-width-230@tablet-only u-letter-spacing-0028@tablet").get_text()
    artista_top_1 = soup.find("span", class_="c-label a-no-trucate a-font-primary-s lrv-u-font-size-14@mobile-max u-line-height-normal@mobile-max u-letter-spacing-0021 lrv-u-display-block a-truncate-ellipsis-2line u-max-width-330 u-max-width-230@tablet-only u-font-size-20@tablet").get_text()
    
    cancion_top1_clean = cancion_top_1.replace("\t","").replace("\n","")
    artista_top1_clean = artista_top_1.replace("\t","").replace("\n","")

    Billboard_Hot_100 = pd.DataFrame({"cancion": [cancion_top1_clean], "artista": [artista_top1_clean]})
    Billboard_Hot_100_1 = pd.DataFrame({"cancion": titulo_cancion, "artista": nombre_artista})
    Billboard_Hot_100_tm = pd.concat([Billboard_Hot_100, Billboard_Hot_100_1], axis=0)

    resultado = Billboard_Hot_100_tm[Billboard_Hot_100_tm['cancion'].str.contains(a, case=False)]

    if not resultado.empty:
        print(resultado)
        b = input('¿Es alguna de estas la canción que buscas? (sí/no): ')
        if b.lower() == 'sí' or b.lower() == 'si' or b.lower() == 'yes':
            print("¡Esta canción está en el Hot 100 de Billboard!")
        else:
            print("No hemos sido capaces de encontrar tu canción.")
    else:
        print("No hemos sido capaces de encontrar tu canción.")

billboard_scrap()
```
