```python
data = pd.read_excel('anova_lab_data.xlsx', sheet_name='data_collected')
data
```

```python
## Part 1

#In this activity, we will look at another example. Your task is to understand the problem and write down all the steps to set up ANOVA. After the next lesson, we will ask you to solve this problem using Python. Here are the steps that you would need to work on:
    #- Null hypothesis
    #- Alternate hypothesis
    #- Level of significance
    #- Test statistic
    #- P-value
    #- F table

#1. Contexto
#Estamos analizando un proceso de grabado de plasma con respecto a la variación de la potencia (en vatios) del haz de plasma. El objetivo es determinar si cambiar la potencia del haz de plasma afecta la tasa de grabado.

#2. Hipótesis
#Hipótesis Nula: La tasa media de grabado es la misma para todos los niveles de potencia del haz de plasma.
#Hipótesis Alternativa: Al menos una tasa media de grabado es diferente de las demás.

#3. Nivel de Significancia
#El nivel de significancia se establece típicamente en 0.05. Esto significa que estás dispuesto a aceptar un 5% de probabilidad de rechazar incorrectamente la hipótesis nula.

#4. Test statistic
#F=Media de los Cuadrados Entre Grupos / Media de los Cuadrados Dentro de los Grupos

#Media de los Cuadrados Entre Grupos: Mide la variación entre las diferentes medias de los grupos.
#Media de los Cuadrados Dentro de los Grupos: Mide la variación dentro de cada grupo.

#5. P-value
#El valor p se calcula en función del test statistic y los grados de libertad. Ayuda a determinar si se debe rechazar la hipótesis nula. Si el valor p es menor que el nivel de significancia, se rechaza la hipótesis nula.

#6. F table

#Calcular las medias de los grupos y la media general.
#Calcular la suma de cuadrados entre grupos.
#Calcular la suma de cuadrados dentro de los grupos.
#Calcular las medias de los cuadrados
#Calcular el estadístico F.
#Determinar el valor p usando la distribución F con los grados de libertad.
```

```python
## Part 2

#- In this section, use Python to conduct ANOVA.
#- What conclusions can you draw from the experiment and why

data.columns = [col.replace(' ', '_') for col in data.columns]
data
```

```python
# Ajustar el modelo ANOVA
model = ols('Etching_Rate ~ C(Power_)', data=data).fit()
table = sm.stats.anova_lm(model)

# Mostrar la tabla ANOVA
print(table)
```

```python
#¿Qué conclusiones obtuvimos?

#Los resultados del ANOVA muestran un valor p de la prueba F inferior a 0,05. Esto significa que podemos rechazar la Hipotesis nula. Es decir, hay diferencias significativas entre la tasa media de grabado para al menos un nivel de potencia.
```
